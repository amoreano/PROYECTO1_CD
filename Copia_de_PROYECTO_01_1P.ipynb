```python
import pandas as pd
```

Se crea los datos de la DataFrame

```python
fch=["01-03-2022","zapato",2,80]
nom=["02-03-2022","camisa",3,20]
prd=["04-03-2022","pantalón",4,30]
cnt=["06-03-2022","blusa",1,28]
fld=["07-03-2022","falda",2,40]
ctr=["09-03-2022","cartera",1,120]
```

Se crea una lista con las variables

```python
lista_pro=[fch,nom,prd,cnt,fld,ctr]
```

Se crea una lista con los encabezados del DataFrame

```python
df_prod=pd.DataFrame(lista_pro,columns=["Fecha","Nombre","Cantidad","Precio"])
```

Se ejecuta el código

```python
df_prod
```

Se haya el Total de venta multiplicando la Cantidad por el Precio

```python
df_prod['Total'] = df_prod['Cantidad'] * df_prod['Precio']
```

```python
df_prod
```

Se muestra la venta de los últimos días tambien se puede codificar con df_prod.tail()

```python
df_prod.tail(5)
```

Se obtiene el 5% de los Totales

```python
df_prod["Descuento 5%"]=df_prod["Total"]*0.05
```

```python
df_prod
```

Y se descuenta el 5% del Total resultando un Total Final

```python
df_prod["Tota Final"]=df_prod["Total"]-df_prod["Descuento 5%"]
df_prod
```
