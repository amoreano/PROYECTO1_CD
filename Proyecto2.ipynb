***Generar 5 nuevas variables con los siguientes criterios***

```python
def mio(df,col1,col2):
    return np.log10(df[col1]/df[col2])
df["nuevo1"]=mio(df,"Millas_Por_Galon","Cilindros")
df                 
```

```python
import math 
from math import exp
import numpy as np 
```

```python
def mio1(df,col1,col2):
    return np.sqrt(df[col1]*np.exp(df[col2])/100)
df["nuevo2"]=mio1(df,"Millas_Por_Galon","Cilindros")
df 
```

```python
def mio(df,col1,col2):
    return 1/np.log(df[col1]/df[col2])*100
df["nuevo3"]=mio(df,"Millas_Por_Galon","Cilindros")
df
```

```python
from matplotlib import style
import seaborn as sns
import matplotlib.pyplot as plt
from pylab import savefig
```

```python
def grafico_num(columnas,data,x):
  for i,colum in enumerate(columnas):
    fig,ax=plt.subplots(figsize=(6,3.84))
    sns.boxplot(
        x=x,
        y=colum,
        data=data,
        )
    ax.set_title(colum, fontsize=10, fontweight="bold")
    ax.tick_params(labelsize=8)
    ax.set_xlabel("Año")
```

```python
x=RiesgoNum.Año
grafico_num(columnas,RiesgoNum,x)
```
